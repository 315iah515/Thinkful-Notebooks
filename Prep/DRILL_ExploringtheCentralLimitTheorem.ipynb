```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ttest_ind

%matplotlib inline
```

# Central Limit Theorem Experiment

The following binomial distribution is used as the population from which samples are created from.

```python
pop1 = np.random.binomial(10, 0.1, 10000)
mean = np.mean(pop1)
sd = np.std(pop1, ddof=1)
print('population mean : %f' % mean)
print('population standard deviation : %f' % sd)
plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.legend(loc='upper right') 
plt.show()
```

# Experment One

Use a the sample szie of 1000, then calculate the means and standard deviations for your sample and create histograms for each. Repeat this again, decreasing the size of the sample to 20

### Expectations : 

I would expect that with a sample size of 1000 from the binomial distribution that the histogram would appear like a normal distribution and that the mean and standard deviation would be close to that of the population.  With a sample size of 20 these expectations would no long hold true, the shape of the histogram may not appear normal in appearance . I would also expect that mean and standard deviation to deviate far from the populations.  

```python
sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop1, 20, replace=True)

plt.hist(sample1, label='sample 1')
plt.legend(loc='upper right') 
plt.show()
mean = np.mean(sample1)
sd = np.std(sample1, ddof=1)
print('sample 1 mean : %f' % mean)
print('sample 1 standard deviation : %f' % sd)


plt.hist(sample2, color='green', label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
mean = np.mean(sample2)
sd = np.std(sample2, ddof=1)
print('sample 2 mean : %f' % mean)
print('sample 2 standard deviation : %f' % sd)
```

# Experiment Two

Binomial population value p changed from .1 to 0.3, then take new samples and compute the t-statistic and p-value. Then change the population value p to 0.4, and repeat the creation of the t-statistic and p-value.

### Expectations :

I would expect that the sample with 1000 observations would not significantly differ from its population. Yet I would expect that the sample with 20 observations to be significantly different from  its population  or least have a greater t value signifying the difference in the sample is due to  variance in the sample. I do know what a change in the value p would have on the  binomial distribution

#### Part A

```python
pop1 = np.random.binomial(10, 0.3, 10000)
mean = np.mean(pop1)
sd = np.std(pop1, ddof=1)
print('population mean : %f' % mean)
print('population standard deviation : %f' % sd)
plt.hist(pop1, alpha=0.5, label='Population 1 with p of 0.3') 
plt.legend(loc='upper right') 
plt.show()
```

```python
sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop1, 20, replace=True)

plt.hist(sample1, label='sample 1')
plt.legend(loc='upper right') 
plt.show()
mean = np.mean(sample1)
sd = np.std(sample1, ddof=1)
print('sample 1 mean : %f' % mean)
print('sample 1 standard deviation : %f' % sd)


plt.hist(sample2, color='green', label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
mean = np.mean(sample2)
sd = np.std(sample2, ddof=1)
print('sample 2 mean : %f' % mean)
print('sample 2 standard deviation : %f' % sd)
```

```python
diff = sample2.mean() - sample1.mean()
print('Difference between the two sample means : %f ' % diff)
size = np.array([len(sample1), len(sample2)])
sd = np.array([sample1.std(), sample2.std()])

se = sum(sd ** 2 / size) ** 0.5
print('Standard error %f :' % se)
print(diff/se)
print(ttest_ind(sample2, sample1, equal_var=False))
```

#### Part B

```python
pop1 = np.random.binomial(10, 0.4, 10000)
mean = np.mean(pop1)
sd = np.std(pop1, ddof=1)
print('population mean : %f' % mean)
print('population standard deviation : %f' % sd)
plt.hist(pop1, alpha=0.5, label='Population 1 with p of 0.4') 
plt.legend(loc='upper right') 
plt.show()
```

```python
sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop1, 20, replace=True)

plt.hist(sample1, label='sample 1')
plt.legend(loc='upper right') 
plt.show()
mean = np.mean(sample1)
sd = np.std(sample1, ddof=1)
print('sample 1 mean : %f' % mean)
print('sample 1 standard deviation : %f' % sd)


plt.hist(sample2, color='green', label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
mean = np.mean(sample2)
sd = np.std(sample2, ddof=1)
print('sample 2 mean : %f' % mean)
print('sample 2 standard deviation : %f' % sd)
```

```python
diff = sample2.mean() - sample1.mean()
print('Difference between the two sample means : %f ' % diff)
size = np.array([len(sample1), len(sample2)])
sd = np.array([sample1.std(), sample2.std()])

se = sum(sd ** 2 / size) ** 0.5
print('Standard error %f :' % se)
print(diff/se)
print(ttest_ind(sample2, sample1, equal_var=False))
```

It would appear as the population distribution becomes more Gaussian shaped even those samples with observations less than 30 can still have a mean and standard deviation very similar to the population from which the events were observed. 

## Experiment Three

Observing the Central Limit Theorem in action. In the first histogram 60 samples are averaged and displayed. In the second histogram 60 samples of 10 observations are averaged and displayed. All samples are selected from a gamma distribution.

### Expectations :

The histogram of 60 different sample means comprised of 200 observations should look like a normal distribution. The second histogram comprised of 60 means of 10 observations each should look less like a normal distribution. 

```python
pop1 = np.random.gamma(1 ,2, 100000)
mean = np.mean(pop1)
sd = np.std(pop1, ddof=1)
print('population mean : %f' % mean)
print('population standard deviation : %f' % sd)
plt.hist(pop1, alpha=0.5, label='Population 1 Gamma') 
plt.legend(loc='upper right') 
plt.show()
```

```python
a_means = np.zeros(60)


for i in range(60):
    a_means[i] = np.mean(np.random.choice(pop1, 200, replace=True))
    
plt.hist(a_means, label='means from 60 samples each with 200 observations')
plt.legend(loc='upper right') 
plt.show()
mean = np.mean(a_means)
sd = np.std(a_means, ddof=1)
print('samples A mean : %f' % mean)
print('samples A standard deviation : %f' % sd)
```

```python
b_means = np.zeros(60)


for i in range(60):
    b_means[i] = np.mean(np.random.choice(pop1, 10, replace=True))
    
plt.hist(b_means, label='means from 60 samples each with 10 observations')
plt.legend(loc='upper right') 
plt.show()
mean = np.mean(a_means)
sd = np.std(b_means, ddof=1)
print('samples B mean : %f' % mean)
print('samples B standard deviation : %f' % sd)
```
