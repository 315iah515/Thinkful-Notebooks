# Central Limit Theorem Experiment

The following binomial distribution is used as the population from which samples are created from.

```python
pop1 = np.random.binomial(10, 0.2, 10000)
mean = np.mean(pop1)
sd = np.std(pop1, ddof=1)
print('population mean : %f' % mean)
print('population standard deviation : %f' % sd)
plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.legend(loc='upper right') 
plt.show()
```

# Experment One

Use a the sample szie of 1000, then calculate the means and standard deviations for your sample and create histograms for each. Repeat this again, decreasing the size of the sample to 20

### Expectations : 

I would expect that with a sample size of 1000 from the binomial distribution that the histogram would appear like a normal distribution and that the mean and standard deviation would be close to that of the population.  With a sample size of 20 these expectations would no long hold true, the shape of the histogram may not appear normal in appearance . I would also expect that mean and standard deviation to deviate far from the populations.  

```python
sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop1, 20, replace=True)

plt.hist(sample1, label='sample 1')
plt.legend(loc='upper right') 
plt.show()
mean = np.mean(sample1)
sd = np.std(sample1, ddof=1)
print('sample 1 mean : %f' % mean)
print('sample 1 standard deviation : %f' % sd)


plt.hist(sample2, color='green', label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
mean = np.mean(sample2)
sd = np.std(sample2, ddof=1)
print('sample 2 mean : %f' % mean)
print('sample 2 standard deviation : %f' % sd)
```
