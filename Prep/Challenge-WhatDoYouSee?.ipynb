# Plotting data from the 2016 US Presidential election 

The number of those voters voting for Trump and their median 2013 household income for every one of the 99 Iowa Counties are displayed in various plots. 
Data obtained from Kaggle at https://www.kaggle.com/joelwilson/2012-2016-presidential-elections

```python
df = pd.read_csv('votes.csv',  usecols=[4,7,10,11,55])
```

```python
JustIowaDf = df.loc[df['state_abbr'].isin(['IA'])]
JustIowaDf['scaled_percent'] = JustIowaDf['Trump'] * 100
JustIowaDf
```

```python
plt.hist(JustIowaDf['votes_gop_2016'], color='blue', bins=10)
plt.title('Iowans voting for Trump')
plt.xlabel('Number of votes')
plt.ylabel('County Count')
plt.show()
```

The above histogram shows the total number of votes cast for the 99 counties within Iowa.  From the diagram one can tell that most counties are small in population and that the number of votes for Trump is under 10,000. 

```python
plt.hist(JustIowaDf['INC110213'], color='green', bins=10)
plt.title('Median Household income from 2013 Census')
plt.xlabel('Median Household income in dollars')
plt.ylabel('County Count')
plt.show()
```

As of 2013 most households in Iowa were making $45,000 to $55,000 a year. I suspect the outliers are those counties around Des Monies, Dallas and Polk.

```python
plt.boxplot(JustIowaDf['INC110213'].values)
plt.ylabel('2013 Median Household Income')
plt.title('Display of Median Income Distribution')
plt.show()
```

Above diagram is a box plot of the Iowa median household incomes in dollars for 2013.  There a two outliers beyond the outer fence or beyond the 3rd quartile.  The median looks to be just below $50,000 dollars.

```python
plt.scatter(y=JustIowaDf['scaled_percent'], x=JustIowaDf['INC110213'], color='purple')
plt.xlabel('2013 Median Hosehold income')
plt.ylabel('Percentage of voters in favor of Trump')
plt.title('Median income VS County percentage for Trump')
plt.show()
```

There would appear that there is no association between those Iowa counties having a higher percentage of votes cast of Trump and the Median household income. The data might contain a slight negative direction but the form of data would suggest a weak association at best.
